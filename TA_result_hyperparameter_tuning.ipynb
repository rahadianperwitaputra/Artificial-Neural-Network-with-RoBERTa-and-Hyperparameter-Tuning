##Import Library

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dense, Dropout
```

```python
df = pd.read_csv('socialfeature_liwc_roberta.csv')
df.head()
```

```python
df.drop(['id','akun'], axis=1, inplace=True)
```

##Alokasi data X dan y

```python
X = df.iloc[:, 1:]
```

```python
y = df.iloc[: , :1]
```

```python
X
```

```python
y
```

##Label encoder terhadap kolom 'label' 

Keterangan label:

Agreeableness = 0

Conscientiousness = 1

Extraversion = 2

Neuroticism = 3

Openness = 4





```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X.iloc[:, 0] = le.fit_transform(X.iloc[:, 0])
y.iloc[:, 0] = le.fit_transform(y.iloc[:, 0])
```

```python
y[y['label']==0].shape
```

```python
X
```

##Proses Scalling menggunakan MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
num_cols = []
for col in X.columns:
  num_cols.append(col)
```

```python
scaler = MinMaxScaler()
```

```python
df_copy = X
df_copy[num_cols] = scaler.fit_transform(df_copy[num_cols])
df_copy.head()
```

```python
y.head(20)
```

##SMOTE

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(df_copy, y)
```

##Split Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size = 0.2, random_state = 42)
```

##Model ANN

```python
from keras.models import Sequential
from keras.layers import Dense

import tensorflow as tf


y_train_cat = tf.keras.utils.to_categorical(y_res, num_classes=5)
y_test_cat = tf.keras.utils.to_categorical(y_test, num_classes=5)


classifier = Sequential()
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu', input_dim = 988))
classifier.add(Dropout(rate=0.2))

# # Adding the second hidden layer
# classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))

# # Adding the third hidden layer
# classifier.add(Dense(units = 140, kernel_initializer = 'uniform', activation = 'relu'))

# # Adding the fourth hidden layer
# classifier.add(Dense(units = 120, kernel_initializer = 'uniform', activation = 'relu'))

# # Adding the fifth hidden layer
# classifier.add(Dense(units = 100, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN | means applying SGD on the whole ANN
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_res, y_train_cat, batch_size = 128, epochs = 100, verbose = 1)

score, acc = classifier.evaluate(X_res, y_train_cat,
                            batch_size=128)
print('Train score:', score)
print('Train accuracy:', acc)
# Part 3 - Making predictions and evaluating the model

# Predicting the Test set results
y_pred = classifier.predict(X_test)

score, acc = classifier.evaluate(X_test, y_test_cat,
                            batch_size=128)
print('Test score:', score)
print('Test accuracy:', acc)
```

##Confussion Matrix

```python
y_pred_cat = y_pred.argmax(axis=1)
```

```python
y_pred_cat
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(y_test, y_pred_cat))
print(accuracy_score(y_test, y_pred_cat))
print(classification_report(y_test, y_pred_cat))
```
